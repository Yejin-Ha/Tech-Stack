```python
# 여러 연도 중 2015년도만 기준으로 비교하기로 결정
# 2015년도 연도만 조회하여 새로운 테이블 생성

s_merge_total2015 = s_merge_total.loc[s_merge_total['연도'] == 2015]
s_merge_total2015

# 2015년도 10월, 11월 12월 데이터가 있는 것 확인
# 해당 월의 평균값을 내서 2015년도 값으로 비교하기로 결정
```

```python
# 각 지역별(월별) 2015년도 분양가격 평균내기
# 10 11 12 의 평균은 11이여서 11월값만 추출 (작업 시작)

s_merge_total2015 = s_merge_total2015.groupby(['지역명'], as_index=False).mean()
s_merge_total2015
```

```python
# 무의미한 '월' 열은 삭제

del s_merge_total2015['월']
```

```python
# 최종 테이블 확인

s_merge_total2015
```

```python
# 서울과 다른 지역과의 가격 차이 비교를 위해 서울 분양가(19725 로 통일하여 비교가격을 확인할려고 새롭게 추가) 열 추가


s_merge_total2015['서울분양가'] = 19725.2
s_merge_total2015
```

```python
# 지역분양가격에서 서울분양가를 뺀 새로운 열'서울과의 차액'을 생성

s_merge_total2015['서울과 차액'] = s_merge_total2015['분양가격'] - s_merge_total2015['서울분양가']
s_merge_total2015

# 서울과의 차액의 min값과 max값이 궁금하여 해당 row를 추출하기로 결정
```

```python
# 서울과 차액의 min값 조회(1) # 서울과 차액이 제일 큰 곳 (음수값임)

result_min = s_merge_total2015['서울과 차액'].min()
result_min

# 서울과의 차액이 가장 큰 지역명이 궁금해서 해당 row 조회하기로 결정
```

```python
# 서울과 차액의 min값 row 조회(2)

s_merge_total2015.loc[s_merge_total2015['서울과 차액'] == result_min]

# 조회결과 전남인 것을 확인
```

```python
# 서울과 차액의 max값 조회(1) (최소 값)
result_max = s_merge_total2015['서울과 차액'].max()
result_max

s_merge_total2015[s_merge_total2015['서울과 차액']==result_max]
    
# 분석 결과 서울이 제일 집값이 비싼 것을 확인
```
