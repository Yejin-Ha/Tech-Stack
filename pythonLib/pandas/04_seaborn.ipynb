# 산점도 그래프

변수(피처)들 의미
1. total_bill - 팁을 준 사람들이 레스토랑에서 계산한 금액
2. tip - 팁 금액
3. sex - 성별
4. smoker - 흡연 여부
5. day - 요일
6. time - 식사 계산 시간
7. size - 식사 인원

````python
# 레스토랑 팁 데이터 셋 활용
tips = sns.load_dataset('tips')
tips.head()
````

- 범주형 데이터 = 범위가 지정된 데이터(ex. 혈액형, 흡연자/비흡연자 ...)
- relplot : 산점도 그래프
```
sns.relplot(x='total_bill'   x축의 데이터 col
            y='tip'          y축의 데이터 col
            data=tips        그래프에 표시될 데이터
            hue='smoker'     데이터 중 구분할 데이터 설정, 범주형 데이터에만 적용이 가능하다.
            or hue='size'    size 컬럼으로 데이터의 색을 구분한다.
            style='smoker'   smoker 데이터를 기준으로 모양을 다르게 표시한다. 범주형 데이터만 적용 가능
            palette='Blues'  전체 데이터의 색을 변경한다.(빨간색 -> 파란색)
        )
```

**맨 마지막에 ;를 추가하면 그래프의 주소 정보를 생략한다**

````python
# 그래프에 박스 생성된다
sns.relplot(x='total_bill', y='tip', data=tips) 
````

````python
sns.relplot(x='total_bill', y='tip', data=tips);
````

- set_style() : 그래프의 스타일을 적용한다.
```
set_style(style='darkgrid'   배경색을 변경한다.
)
```

````python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips);
````

````python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker');
````

````python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size');
````

````python
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker');
````

````python
# palette 속성을 통해 데이터의 색 바꾸기(빨간색 -> 파란색으로 변경)
sns.set_style(style='darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='smoker', palette='Blues');
````

## catplot

- 요일별 tip이 수행된 데이터를 시각화
- x축을 요일, y축을 tip으로 하는 그래프 생성
- catplot() : 범주형 변수와 수(치)형 변수간 관계 시각화한 그래프
```
sns.catplot(x='day'         x축의 데이터
            y='tip'         y축의 데이터
            data=tips       그래프에 표시할 데이터
            kind='swarm'    데이터를 겹치지 않게 표현하는 속성
                            -> 데이터양이 많을 경우 부적합하다.
            kind='box'      박스 plot 그래프 그리기. 이상치는 점으로 표현된다.
            or kind='boxen' box 그래프랑 같은 내용을 보여주지만 생김새가 다르다.
            
            kind='violin'   바이올린 plot 그래프 그리기
            
            kind='bar'      바 plot 그래프 그리기
            hue='day'       바 plot 그래프의 굵기도 변경 가능하다.
            
            kind='point'    point를 적용하는 범주형 데이터 차트 그리기
```

````python
sns.catplot(x='day', y='tip', data=tips);
````

````python
sns.catplot(x='day', y='tip', data=tips, kind='swarm');
````

````python
sns.catplot(x='day', y='tip', data=tips, kind='box');
````

````python
sns.catplot(x='day', y='tip', data=tips, kind='boxen');
````

````python
sns.catplot(x='day', y='tip', data=tips, kind='violin');
````

````python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar');
````

````python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar', hue='day');
````

````python
sns.catplot(x='day', y='total_bill', data=tips, kind='point');
````

## histplot
- histplot() : 변수에 대한 히스토그램을 표시한다.
```
sns.histplot(data=penguins         그래프에 표시할 데이터
            x='flipper_length_mm'  x축의 데이터
            hue='species'          데이터를 구분할 col명
            multiple='stack'       데이터의 중첩을 피하겠다는 설정
            )
```

````python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack");
````

````python
# stack을 사용하지 않아서 중첩된 모습
sns.histplot(data=penguins, x="flipper_length_mm", hue="species");
````

## heatmap

````python
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
flights = flights_long.pivot("month", "year", "passengers")
flights
````

- heatmap() : 
```
sns.heatmap(datas           그래프에 표시할 데이터
            annot=True      각 데이터의 표현식 표시한다.
            linewidths=.5   데이터 사이의 여백을 표현하는 속성
            ax=ax           생략시 최소 사이즈로 그린다.
            fmt='d'         데이터를 정수로 출력하는 포멧 설정
            

````python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
````

````python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, fmt="d", linewidths=.5, ax=ax);
````

````python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", ax=ax);
````

````python
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5);
````

````python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, linewidths=.5, ax=ax);
````
